```
import pandas as pd
import datetime as dt
from pyspark.sql import SparkSession
import pyspark
```

```
#Definimos la variable del día actual para extraer el año en el bucle de ingesta 
today = dt.datetime.today()

#Creamos el DF de destino
final_df = pd.DataFrame()

#Recorremos los links de github para importar los partidos de los distintos años
for year in range(1968, today.year):
  raw_df = pd.read_csv("https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_"+str(year)+".csv")
  final_df = final_df.append(raw_df, ignore_index=True)

#Convertimos a fecha la columna tourney_date
final_df["tourney_date"] = pd.to_datetime(final_df["tourney_date"], format='%Y%m%d')
```

```
final_df[['surface','winner_seed','winner_entry','winner_hand','winner_ioc','loser_seed','loser_entry','loser_hand','loser_ioc','score']] = final_df[['surface','winner_seed','winner_entry','winner_hand','winner_ioc','loser_seed','loser_entry','loser_hand','loser_ioc','score']].astype(str)
```

```
#Convertimos a DS de Spark
tennis_raw = spark.createDataFrame(final_df)
```
