```
#Cantidad de juegos ganados/perdidos por jugador por tipo de piso   
MySparkSession.sql("""SELECT * FROM 
                   (SELECT winner_name, 'Ganados' AS Partidos, surface,  COUNT (winner_id) AS win
                   FROM tenis_matches WHERE winner_name = 'Radek Stepanek'
                   GROUP BY winner_name, surface) 
                   AS P
                   PIVOT(SUM (win) FOR surface in ('Clay', 'Grass', 'Hard', 'Carpet'))
                   UNION
                   SELECT * FROM 
                   (SELECT loser_name, 'Perdidos' AS Partidos, surface,  COUNT (loser_id) AS lose
                   FROM tenis_matches WHERE loser_name = 'Radek Stepanek'
                   GROUP BY loser_name, surface) 
                   AS P
                   PIVOT(SUM (lose) FOR surface in ('Clay', 'Grass', 'Hard', 'Carpet'))
                   """).show()
```

```
#=====
#Procedimiento para listar todos los jugadores por juegos ganados/perdidos
#La idea es tener un dataframe con los datos por jugador, no por partido
#=====                    


def listaPlayers2(dataFrame):
    ini = datetime.datetime.now()
    print ("Hora inicio: ")
    print (ini.strftime("%Y-%m-%d %H:%M:%S"))
    count = 1
    datosPlayers=pd.DataFrame()
    for data in dataFrame["winner_name"]:
        player1 = data
        #print(player1)
        playersW = MySparkSession.sql("""SELECT tourney_id, tourney_name, draw_size, 
                                      tourney_level, tourney_date, match_num, winner_id as player_id, winner_seed AS seed,  
                                      winner_name as player_name, loser_name as content_name, winner_hand as hand, winner_ht as ht, winner_ioc as ioc, winner_age as age, score, best_of, 
                                      round, minutes, w_ace as ace, w_df as df, w_svpt as svpt, w_1stIn as 1stIn, w_1stWon as 1stWon, w_2ndWon as 2ndWon, w_SvGms as SvGms, 
                                      w_bpSaved as bpSaved, w_bpFaced as bpFaced, winner_rank as rank, winner_rank_points as rank_points, Carpet, Clay, Grass, 
                                      Hard, A, C, D, F, G, M, W_ALT60 as alt1, W_Alt61 as alt2, W_LL as ll, W_PR as pr, W_Q as q, W_SE as se, W_WC as wc, W_L as l, W_R as r, 
                                      W_U as u, Year, '1' AS WIN FROM tenis_matches 
                                      WHERE winner_name =  '"""+player1+"""'""").to_koalas()#toPandas() 
                                      
        playersL = MySparkSession.sql("""SELECT tourney_id, tourney_name, draw_size, 
                                      tourney_level, tourney_date, match_num, loser_id as player_id, loser_seed as seed, 
                                      loser_name as player_name,  winner_name as content_name, loser_hand as hand, loser_ht as ht, loser_ioc as ioc, loser_age as age, score, best_of, 
                                      round, minutes, l_ace as ace, l_df as df, l_svpt as svpt, l_1stIn as 1stIn, l_1stWon as 1stWon, l_2ndWon as 2ndWon, l_SvGms as SvGms, 
                                      l_bpSaved as bpSaved, l_bpFaced as bpFaced, loser_rank as rank, loser_rank_points as rank_points, Carpet, Clay, Grass, 
                                      Hard, A, C, D, F, G, M, L_ALT67 as alt1, L_Alt68 as alt2, L_LL as ll, L_PR as pr, L_Q as q, L_SE as se, L_WC as wc, L_L as l, L_R as r, 
                                      L_U as u, Year, '0' AS WIN FROM tenis_matches 
                                      WHERE loser_name =  '"""+player1+"""'""").to_koalas()#toPandas() 
        
        frames = [playersW, playersL]  
        result = pd.concat(frames)
        datosPlayers= datosPlayers.append(result)
        print(count) 
        count = count + 1
    fin = datetime.datetime.now() 
    print ("Hora inicio: ")
    print (ini.strftime("%Y-%m-%d %H:%M:%S"))
    print ("Hora fin : ")    
    print (fin.strftime("%Y-%m-%d %H:%M:%S"))    
    return datosPlayers   
```
